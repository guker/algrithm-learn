```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

im = cv2.imread('testfile.jpg')


plt.figure(figsize=(20,20))

plt.subplot(2,1,1)
plt.imshow(im[:,:,::-1])


detector = cv2.SIFT()
keypoints = detector.detect(im)

im = cv2.drawKeypoints(im, keypoints)
plt.subplot(2,1,2)
plt.imshow( im[:,:,::-1])
```
